```python
import pandas as pd
bread = pd.read_csv('bread.csv')
bread
```

```python
x = bread.drop(columns=['Bread'])
x
```

```python
y=bread['Bread']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x,y)
prediction=model.predict([[35,0]])
prediction
```

```python
import joblib
joblib.dump(model,'Bread-recommender.joblib')
```
